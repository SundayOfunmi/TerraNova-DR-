```python
import pandas as pd

def calculate_duration(df):
    # 1. Convert columns to datetime objects
    df['incidentBeginDate'] = pd.to_datetime(df['incidentBeginDate'])
    df['incidentEndDate'] = pd.to_datetime(df['incidentEndDate'])

    # 2. Handle missing End Dates (Option: Use Begin Date if missing, resulting in 0 days)
    df['incidentEndDate'] = df['incidentEndDate'].fillna(df['incidentBeginDate'])

    # 3. Calculate the difference in days
    # We add +1 because a disaster that starts and ends on the same day is often counted as a 1-day event
    df['incident_duration_days'] = (df['incidentEndDate'] - df['incidentBeginDate']).dt.days + 1
    
    # 4. Optional: Handle negative values (if data errors exist)
    df['incident_duration_days'] = df['incident_duration_days'].clip(lower=1)
    
    return df
```
